```python
import re

def ratePass(x):
    if x[0]!='0':
        if len(x)>3 and x[3]<='C':
            return 1.0/int(x[0])
        else:
            return -1.0/int(x[0])
    else:
        return 0.0

def encode(s):
    v = re.split(r':',s)
    x = v[1][1:]
    out = 0
    if x=='A+':
        out=4.0
    if x=='A':
        out=4.0
    if x=='A-':
        out=3.7
    if x=='B+':
        out=3.3
    if x=='B':
        out=3.0
    if x=='B-':
        out=2.7
    if x=='C+':
        out=2.3
    if x=='C':
        out=2.0
    if x=='C-':
        out=1.7
    if x=='D+':
        out=1.3
    if x=='D':
        out=1.0
    if x=='D-':
        out=0.7
    if x=='F':
        out=0.0
    if x=='NA':
        out = 2.0
    if x=='CR':
        out=2.0
    #return [int(v[0]),out-2]
    #return int(v[1][0])*out-1-int(v[0])
    return int(v[1][0])*out-2 #incorporate credit data
    
    
    
```
