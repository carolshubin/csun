```python
%matplotlib inline
from __future__ import division
import numpy as np
import pandas as pd
#from sklearn.linear_model import LogisticRegression
import sklearn
from sklearn import metrics
from sklearn.cross_validation import train_test_split, cross_val_score
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('math-redacted.csv')
data.head()
```

```python
data['Graduated'].replace(to_replace=('N','Y'), value=(0,1), inplace=True)
data['Major_Cd'].replace(to_replace=('785845B','785834B'), value=(0,1), inplace=True)
# 785845B math bs
```

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.cross_validation import train_test_split
from sklearn.metrics import classification_report
from sklearn.pipeline import Pipeline
from sklearn.grid_search import GridSearchCV
```

```python
#y = data[data['Graduated']==1]['Major_Cd']
y = data['Graduated']
##y = data['Major_Cd']
X = data[data.columns[6:]]
#X = data[data['Graduated']==1][data.columns[6:]]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
pipeline = Pipeline([
           ('clf', DecisionTreeClassifier(criterion='entropy'))
])
```

```python
parameters = {
           'clf__max_depth': (100, 155, 200),
           'clf__min_samples_split': (1, 2, 3),
           'clf__min_samples_leaf': (1, 2, 3,5)
}
```

```python
grid_search = GridSearchCV(pipeline, parameters, n_jobs=-1,verbose=1, scoring='f1')
grid_search.fit(X_train, y_train)
print 'Best score: %0.3f' % grid_search.best_score_
print 'Best parameters set:'
best_parameters = grid_search.best_estimator_.get_params()
for param_name in sorted(parameters.keys()):
        print '\t%s: %r' % (param_name, best_parameters[param_name])
predictions = grid_search.predict(X_test)
print classification_report(y_test, predictions)
```

```python
confusion_matrix = metrics.confusion_matrix(y_test, predictions)
print(confusion_matrix)
plt.matshow(confusion_matrix)
plt.title('Confusion matrix')
plt.colorbar()
plt.ylabel('True label')
plt.xlabel('Predicted label')
```

```python
from sklearn.ensemble import RandomForestClassifier

pipeline = Pipeline([
           ('clf', RandomForestClassifier(criterion='entropy'))
       ])
parameters = {
           'clf__n_estimators': (5, 10, 20, 50),
           'clf__max_depth': (50, 150, 250),
           'clf__min_samples_split': (1, 2, 3),
           'clf__min_samples_leaf': (1, 2, 3)
}

y = data['Graduated']
##y = data['Major_Cd']
X = data[data.columns[6:]]

X_train, X_test, y_train, y_test = train_test_split(X, y)
grid_search = GridSearchCV(pipeline, parameters, n_jobs=-1,verbose=1, scoring='f1')
grid_search.fit(X_train, y_train)
print 'Best score: %0.3f' % grid_search.best_score_
print 'Best parameters set:'
best_parameters = grid_search.best_estimator_.get_params()
for param_name in sorted(parameters.keys()):
        print '\t%s: %r' % (param_name, best_parameters[param_name])
predictions = grid_search.predict(X_test)
print classification_report(y_test, predictions)

confusion_matrix = metrics.confusion_matrix(y_test, predictions)
print(confusion_matrix)
plt.matshow(confusion_matrix)
plt.title('Confusion matrix')
plt.colorbar()
plt.ylabel('True label')
plt.xlabel('Predicted label')
```
