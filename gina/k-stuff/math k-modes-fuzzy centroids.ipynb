```python
%matplotlib inline
from __future__ import division
import numpy as np
import pandas as pd
#from sklearn.linear_model import LogisticRegression
import sklearn
from sklearn import metrics
from sklearn.cross_validation import train_test_split, cross_val_score
import matplotlib.pyplot as plt
from utilities import printClusters,plotPCA,getDistancesDF,clusterCenters
from scipy.spatial.distance import cdist
from sklearn import metrics
```

```python
data = pd.read_csv('math-redacted.csv')
data.head()
```

```python
data['Graduated'].replace(to_replace=('N','Y'), value=(0,1), inplace=True)
#data['Major_Cd'].replace(to_replace=('785845B','785834B'), value=(0,1), inplace=True)
# 785845B math bs
```

```python
from kmodes import FuzzyCentroidsKModes
"""Fuzzy k-modes clustering algorithm for categorical data.
        Uses fuzzy centroids, following and Kim, D.-W., Lee, K.H., Lee, D.:
        Fuzzy clustering of categorical data using fuzzy centroids, Pattern
        Recognition Letters 25, 1262-1271, 2004.
"""
k=2

kmod=FuzzyCentroidsKModes(k)

X_raw=data[data.columns[6:]]

kmod._perform_clustering(X_raw) #,init_method='Cao'

gradClustersDF=printClusters(data, kmod.clusters.flatten(),k=k)
plotPCA(data[data.columns[6:]],kmod.clusters.flatten(),k=k,norm=False)

```
