# Ternary encoding

```python
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist
import matplotlib.pyplot as plt

X = data[data.columns[cl:]]
K = range(1, 20)
meandistortions = []
for k in K:
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X)
    meandistortions.append(sum(np.min(cdist(data[data.columns[5:]],\
                                            kmeans.cluster_centers_,\
                                            'euclidean'), axis=1)) / data[data.columns[5:]].shape[0])
plt.plot(K, meandistortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Average distortion')
plt.title('Selecting k with the Elbow Method')
plt.show()
```

```python
#DO NOT RUN IT AGAIN


from sklearn import metrics

tests = np.arange(2,10)


for t in tests:
   kmeans_model = KMeans(n_clusters=t).fit(X)
   print "K=",t," silhouette score", metrics.silhouette_score(data[data.columns[cl:]],\
                                                   kmeans_model.labels_,metric='euclidean')
```

```python
k=2
#center=clusterCenters(data, k=k,n_tests=10,init='random')
kmeans_model = KMeans(n_clusters=k, n_init=20, init='random').fit(data[data.columns[cl:]])
gradClustersDF=printClusters(data, kmeans_model.labels_,k=kmeans_model.n_clusters)
plotPCA(data[data.columns[cl:]],kmeans_model.labels_,k=kmeans_model.n_clusters,norm=False)
```

```python
#distancesDF=pd.DataFrame({'class':data.columns[6:],'Cluster0':kmeans_model.cluster_centers_[0],'Cluster1':kmeans_model.cluster_centers_[1]})

distancesDF=getDistancesDF(data.columns[cl:], kmeans_model.cluster_centers_,k=kmeans_model.n_clusters)
distancesDF['Distance']=distancesDF['Cluster0']-distancesDF['Cluster1']
distancesDF.head()
```

```python
plt.plot(distancesDF.sort('Distance', ascending=True)['Distance'])
print list(distancesDF.sort('Distance', ascending=True)['class'])
```

# One Hot Part

```python

encoder=sklearn.preprocessing.OneHotEncoder(n_values=3)
X=data[data.columns[cl:]]
X_pos=X+1
encoder.fit(X_pos)
```

```python
X_enc = encoder.transform(X_pos).toarray()
K = range(1, 20)
meandistortions = []
for k in K:
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X_enc)
    meandistortions.append(sum(np.min(cdist(X_enc, kmeans.cluster_centers_, 'euclidean'),\
                                      axis=1)) / X_enc.shape[0])
plt.plot(K, meandistortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Average distortion')
plt.title('Selecting k with the Elbow Method')
plt.show()
```

```python
tests = np.arange(2,6)


for t in tests:
   kmeans_model = KMeans(n_clusters=t).fit(X_enc)
   print "K=",t," silhouette score", metrics.silhouette_score(X_enc, kmeans_model.labels_,metric='euclidean')

```

```python
kmeans_model = KMeans(n_clusters=2).fit(encoder.transform(X_pos).toarray())
         
columns_onehot=np.asarray([[c+' fail',c+' DNE',c+' pass'] for c in data.columns[cl:]]).flatten()
distancesDF=getDistancesDF(columns_onehot, kmeans_model.cluster_centers_,k=kmeans_model.n_clusters)
gradClustersDF=printClusters(data, kmeans_model.labels_,k=kmeans_model.n_clusters)
plotPCA(data[data.columns[cl:]],kmeans_model.labels_,k=kmeans_model.n_clusters,norm=False)
```
