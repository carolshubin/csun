```python
data2 = pd.read_csv('/Volumes/Data/College/CSUN/math494/csun/chris/psych-redacted.csv')
```

```python
data2['Graduated'].replace(to_replace=('N','Y'), value=(0,1), inplace=True)
data3=data2.replace(to_replace=set(data2[data2.columns[5:]].values.flatten()),\
                                       value=range(len(set(data2[data2.columns[5:]].values.flatten()))))

```

```python
def attempt(x):
    return int(x[0])

attemptsDF=data2[data2.columns[5:]].applymap(attempt)
attemptsDF[data2.columns[:5]]=data2[data2.columns[:5]]
attemptsDF=attemptsDF[data2.columns]
```

```python

def ratePass(x):
    if x[0]!='0':
        if len(x)>3 and x[3]<='C':
            return 1.0/int(x[0])
        else:
            return -1.0/int(x[0])
    else:
        return 0.0
    
    
def rateGPA(x):
    if x[0]!='0' and len(x)>2 and x[1]==':':
        if len(x)>3 and x[3]<='D':
            if x[3]=='D':    
                return 1.0/int(x[0])
            if x[3]=='C':    
                return 2.0/int(x[0])
            if x[3]=='B':    
                return 3.0/int(x[0])
            if x[3]=='A':    
                return 4.0/int(x[0])
        else:
            if len(x)>3 and x[3]=='W':
                return 0.0
            else: 
                return -1.0/int(x[0])
    else:
        return 0.0

ratedDF=data2[data2.columns[5:]].applymap(ratePass)
ratedDF[data2.columns[:5]]=data2[data2.columns[:5]]
ratedDF=ratedDF[data2.columns]

gpaDF=data2[data2.columns[5:]].applymap(rateGPA)
gpaDF[data2.columns[:5]]=data2[data2.columns[:5]]
gpaDF=gpaDF[data2.columns]
gpaDF['MATH140'][pd.isnull(gpaDF['MATH140'])==True]
```

```python
from sklearn.linear_model import LogisticRegression
import sklearn
from sklearn import metrics
from sklearn.cross_validation import train_test_split, cross_val_score
import matplotlib.pyplot as plt

X_raw=ratedDF[ratedDF.columns[5:]]
X_train_ml, X_test_ml, y_train_ml, y_test_ml = train_test_split(X_raw, ratedDF['Graduated'],test_size=0.2)

'''
X_train=data['Total_Cr'].values
X_test=data['Total_Cr'].values
y_train=data['Graduated'].values
y_test=data['Graduated'].values
'''

classifier = LogisticRegression()
classifier.fit(X_train_ml, y_train_ml)
y_pred_ml = classifier.predict(X_test_ml)

confusion_matrix = metrics.confusion_matrix(y_test_ml, y_pred_ml)
print(confusion_matrix)
plt.matshow(confusion_matrix)
plt.title('Confusion matrix')
plt.colorbar()
plt.ylabel('True label')
plt.xlabel('Predicted label')

print(metrics.classification_report(y_test_ml, y_pred_ml))

coef_list=classifier.coef_
```
