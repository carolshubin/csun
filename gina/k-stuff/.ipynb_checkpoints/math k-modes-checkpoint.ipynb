```python
%matplotlib inline
from __future__ import division
import numpy as np
import pandas as pd
#from sklearn.linear_model import LogisticRegression
import sklearn
from sklearn import metrics
from sklearn.cross_validation import train_test_split, cross_val_score
import matplotlib.pyplot as plt
from utilities import printClusters,plotPCA,getDistancesDF,clusterCenters
from scipy.spatial.distance import cdist
from sklearn import metrics
```

```python
data = pd.read_csv('math-redacted.csv')
data.head()
```

```python
data['Graduated'].replace(to_replace=('N','Y'), value=(0,1), inplace=True)
#data['Major_Cd'].replace(to_replace=('785845B','785834B'), value=(0,1), inplace=True)
# 785845B math bs
```

```python
from kmodes import KModes
X_raw=data[data.columns[6:]]
k=5

def getBestCluster(model, X, k, c=10):
    cost=np.inf
    kmod=None
    for i in range(c):
        kmod_test=model(k)
        kmod_test._perform_clustering(X, verbose=0)
        if cost>kmod_test.cost:
            kmod=kmod_test
    return kmod
kmod=getBestCluster(KModes, X_raw, k)
```

```python
gradClustersDF=printClusters(data, kmod.clusters.flatten(),k=k)
plotPCA(data[data.columns[6:]],kmod.clusters.flatten(),k=k,norm=False)
```

```python
distancesDF=getDistancesDF(data.columns[6:], kmod.centroids,k=k)
distancesDF['Distance']=distancesDF['Cluster0']-distancesDF['Cluster1']
distancesDF.head()

#kmod.clusters.flatten()
```

```python
X = data[data.columns[6:]]
K = range(2,20 )
meandistortions = []
for k in K:
    #km = KModes(k)
    #km._perform_clustering(X, verbose=0)
    km=getBestCluster(KModes, X, k,c=30)
    meandistortions.append(sum(np.min(cdist(X, km.centroids, 'hamming'), axis=1)) )
plt.plot(K, meandistortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Average distortion')
plt.title('Selecting k with the Elbow Method')
plt.show()
```

```python
tests = np.arange(2,10)

for t in tests:
    #km = KModes(t)
    #km._perform_clustering(X, verbose=0)
    km=getBestCluster(KModes, X, t)
    print "K=",t," silhouette score", metrics.silhouette_score(X, km.clusters.flatten(),metric='hamming')
```

```python
from kmodes import FuzzyKModes
"""Fuzzy k-modes clustering algorithm for categorical data.
        Uses traditional, hard centroids, following Huang, Z., Ng, M.K.:
        A fuzzy k-modes algorithm for clustering categorical data,
        IEEE Transactions on Fuzzy Systems 7(4), 1999.
        """
k=2
kmod=FuzzyKModes(k)

X_raw=data[data.columns[6:]]

kmod._perform_clustering(X_raw, verbose=0) #,init_method='Cao'

gradClustersDF=printClusters(data, kmod.clusters.flatten(),k=k)
plotPCA(data[data.columns[6:]],kmod.clusters.flatten(),k=k,norm=False)

```

```python
X = data[data.columns[6:]]
K = range(2,20)
meandistortions = []
for k in K:
    #km = FuzzyKModes(k)
    #km._perform_clustering(X, verbose=0)
    km=getBestCluster(FuzzyKModes, X, k)
    meandistortions.append(sum(np.min(cdist(X, km.centroids, 'hamming'), axis=1)) )
plt.plot(K, meandistortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Average distortion')
plt.title('Selecting k with the Elbow Method')
plt.show()
```

```python
from sklearn.cluster.k_modes_ import KModes as KM

k=2
#center=clusterCenters(data[data.columns[6:]], k=k,n_tests=10,init='random')
kmodes_model = KM(n_clusters=k, n_init=20, init='random').fit(data[data.columns[6:]])
gradClustersDF=printClusters(data, kmodes_model.labels_,k=kmodes_model.n_clusters)
plotPCA(data[data.columns[6:]],kmodes_model.labels_,k=kmodes_model.n_clusters,norm=False)
```

```python
distancesDF=getDistancesDF(data.columns[6:], kmodes_model.cluster_centers_,k=k)
distancesDF['Distance']=distancesDF['Cluster0']-distancesDF['Cluster1']
distancesDF.sort('Distance', ascending=False).head(10)
#distancesDF.sort('Distance', ascending=False)[(distancesDF.sort('Distance', ascending=False)['Distance']>0.3)|(distancesDF.sort('Distance', ascending=False)['Distance']<-0.2)]
```

```python
X = data[data.columns[6:]]
K = range(1, 20)
meandistortions = []
for k in K:
    km = KM(n_clusters=k)
    km.fit(X)
    meandistortions.append(sum(np.min(cdist(X, km.cluster_centers_, 'hamming'), axis=1)) )
plt.plot(K, meandistortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Average distortion')
plt.title('Selecting k with the Elbow Method')
plt.show()
```

```python
tests = np.arange(2,10)

for t in tests:
   km = KM(n_clusters=t).fit(X)
   print "K=",t," silhouette score", metrics.silhouette_score(X, km.labels_,metric='hamming')

```
