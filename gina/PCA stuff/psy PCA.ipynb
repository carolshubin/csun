```python
%matplotlib inline
from __future__ import division
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
import sklearn
from sklearn import metrics
from sklearn.cross_validation import train_test_split, cross_val_score
import matplotlib.pyplot as plt
#from sklearn_pandas import DataFrameMapper, cross_val_score
```

```python
data = pd.read_csv('psych-redacted.csv')
data.head()

```

```python
data['Graduated'].replace(to_replace=('N','Y'), value=(0,1), inplace=True)
```

```python
#using PCA to reduce dimention of vertor space and plot it in 2D
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
X_raw=data[data.columns[5:]]
X_raw_norm= (X_raw - X_raw.mean())# / (X_raw.max() - X_raw.min())
reduced_X = pca.fit_transform(X_raw_norm)
y=data['Graduated']

red_x, red_y = [], []
blue_x, blue_y = [], []
green_x, green_y = [], []
for i in range(len(reduced_X)):
    if y[i] == 0:
        red_x.append(reduced_X[i][0])
        red_y.append(reduced_X[i][1])
    elif y[i] == 1:
        blue_x.append(reduced_X[i][0])
        blue_y.append(reduced_X[i][1])
    else:
        green_x.append(reduced_X[i][0])
        green_y.append(reduced_X[i][1])
plt.scatter(red_x, red_y, c='r', marker='x')
plt.scatter(blue_x, blue_y, c='b', marker='D')
plt.scatter(green_x, green_y, c='g', marker='.')
plt.show()
```

```python
dfPC1=pd.DataFrame({'class':data.columns[5:], 'eig_comp':pca.components_[0]})
dfPC2=pd.DataFrame({'class':data.columns[5:], 'eig_comp':pca.components_[1]})
```
