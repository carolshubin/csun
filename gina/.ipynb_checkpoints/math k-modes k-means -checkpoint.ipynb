```python
%matplotlib inline
from __future__ import division
import numpy as np
import pandas as pd
#from sklearn.linear_model import LogisticRegression
import sklearn
from sklearn import metrics
from sklearn.cross_validation import train_test_split, cross_val_score
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('math-redacted.csv')
data.head()
```

```python
data['Graduated'].replace(to_replace=('N','Y'), value=(0,1), inplace=True)
#data['Major_Cd'].replace(to_replace=('785845B','785834B'), value=(0,1), inplace=True)
# 785845B math bs
```

```python
from kmodes import KModes

kmod=KModes(4)

X_raw=data[data.columns[6:]]

kmod._perform_clustering(X_raw) #,init_method='Cao'


gradClustersDF=pd.DataFrame({'Graduated':data['Graduated'], 'ClusterLbl':kmod.clusters.flatten()})
print 'total graduates', len(gradClustersDF[gradClustersDF['Graduated']==1])
print '0 cluster', gradClustersDF[(gradClustersDF['Graduated']==1)&(gradClustersDF['ClusterLbl']==0)]['ClusterLbl'].count()
print '1 cluster', gradClustersDF[(gradClustersDF['Graduated']==1)&(gradClustersDF['ClusterLbl']==1)]['ClusterLbl'].count()
print '2 cluster', gradClustersDF[(gradClustersDF['Graduated']==1)&(gradClustersDF['ClusterLbl']==2)]['ClusterLbl'].count()
print '3 cluster', gradClustersDF[(gradClustersDF['Graduated']==1)&(gradClustersDF['ClusterLbl']==3)]['ClusterLbl'].count()


print 'total non-graduates', len(gradClustersDF[gradClustersDF['Graduated']==0])
print '0 cluster', gradClustersDF[(gradClustersDF['Graduated']==0)&(gradClustersDF['ClusterLbl']==0)]['ClusterLbl'].count()
print '1 cluster', gradClustersDF[(gradClustersDF['Graduated']==0)&(gradClustersDF['ClusterLbl']==1)]['ClusterLbl'].count()
print '2 cluster', gradClustersDF[(gradClustersDF['Graduated']==0)&(gradClustersDF['ClusterLbl']==2)]['ClusterLbl'].count()
print '3 cluster', gradClustersDF[(gradClustersDF['Graduated']==0)&(gradClustersDF['ClusterLbl']==3)]['ClusterLbl'].count()



from sklearn.decomposition import PCA
pca = PCA(n_components=2)
X_raw=data[data.columns[6:]]
normalizer=sklearn.preprocessing.Normalizer()
#X_raw_norm=normalizer.fit_transform(X_raw.transpose().values.astype('float64')).transpose()
X_raw_norm= (X_raw - X_raw.mean()) # / (X_raw.max() - X_raw.min())
reduced_X = pca.fit_transform(X_raw_norm)
y=data['Graduated']

red_x, red_y = [], []
blue_x, blue_y = [], []
green_x, green_y = [], []
purple_x, purple_y = [], []
for i in range(len(reduced_X)):
    if kmod.clusters.flatten()[i]==0: #red
        red_x.append(reduced_X[i][0])
        red_y.append(reduced_X[i][1])
    elif kmod.clusters.flatten()[i] == 1: #blue
        blue_x.append(reduced_X[i][0])
        blue_y.append(reduced_X[i][1])
    elif kmod.clusters.flatten()[i] == 2:
        green_x.append(reduced_X[i][0])
        green_y.append(reduced_X[i][1])
    elif kmod.clusters.flatten()[i] == 3:
        purple_x.append(reduced_X[i][0])
        purple_y.append(reduced_X[i][1])
plt.scatter(red_x, red_y, c='r', marker='x')
plt.scatter(blue_x, blue_y, c='b', marker='D')
plt.scatter(green_x, green_y, c='g', marker='.')
plt.scatter(purple_x, purple_y, c='m', marker='+')
plt.show()
```

```python
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist
import matplotlib.pyplot as plt

X = data[data.columns[6:]]
K = range(1, 20)
meandistortions = []
for k in K:
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X)
    meandistortions.append(sum(np.min(cdist(X, kmeans.cluster_centers_, 'euclidean'), axis=1)) / X.shape[0])
plt.plot(K, meandistortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Average distortion')
plt.title('Selecting k with the Elbow Method')
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn import metrics

tests = np.arange(2,10)


for t in tests:
   kmeans_model = KMeans(n_clusters=t).fit(X)
   print "K=",t," silhouette score", metrics.silhouette_score(X, kmeans_model.labels_,metric='euclidean')
```
