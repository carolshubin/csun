```python
%matplotlib inline
from __future__ import division
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import re
import sklearn

from sklearn import metrics
from sklearn.cross_validation import train_test_split, cross_val_score
from sklearn import metrics
from sklearn.metrics import roc_curve, auc
from sklearn.linear_model import LogisticRegression
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist

def getGradsClusterDF(df, cluster_list):
    return pd.DataFrame({'Graduated':df['Graduated'],'ClusterLbl':cluster_list})
#x =getGradsClusterDF(data, kmeans_model.labels_)
#x.head()

def printContingencyTable(y,ypred,labels):
    confusion_matrix = metrics.confusion_matrix(y, ypred)
    plt.matshow(confusion_matrix)
    plt.title('Confusion matrix')
    plt.colorbar()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.show()
    print pd.crosstab(y,ypred,rownames=[labels[0]],colnames=[labels[1]])

def plotROC(y_test,y_pred):
    #y_pred can consist of Boolean labels or probabilities, w/ the obvious implications for curve character
    false_positive_rate, true_positive_rate, thresholds = roc_curve(y_test,y_pred)
    roc_auc = auc(false_positive_rate, true_positive_rate)
    plt.title('Receiver Operating Characteristic')
    plt.plot(false_positive_rate, true_positive_rate, 'b', label='AUC = %0.2f' %roc_auc)
    plt.legend(loc='lower right')
    plt.plot([0, 1], [0, 1], 'r--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.0])
    plt.ylabel('False Negative Rate')
    plt.xlabel('True Positive Rate')
    plt.show()
```

```python
#
#Load/encode data
#
data = pd.read_csv('economics_redacted.csv')
data['Graduated'].replace(to_replace=('N','Y'), value=(0,1), inplace=True)

print data.columns

notClassStrings = ['Major_Cd','N_Terms','Graduated','Transfer_Cr','Inst_Cr']
newset = data.columns - notClassStrings

classIndexStart = list(data.columns).index(newset[0])
classIndexEnd = len(data.columns)-1

def encode(s):
    v = re.split(r':',s)
    x = v[1][1:]
    out = 0.0
    if x=='A+':
        out=4.0
    if x=='A':
        out=4.0
    if x=='A-':
        out=3.7
    if x=='B+':
        out=3.3
    if x=='B':
        out=3.0
    if x=='B-':
        out=2.7
    if x=='C+':
        out=2.3
    if x=='C':
        out=2.0
    if x=='C-':
        out=1.7
    if x=='D+':
        out=1.3
    if x=='D':
        out=1.0
    if x=='D-':
        out=0.7
    if x=='F':
        out=0.0
    if x=='NA':
        out = 0.0
    if x=='CR':
        out=2.0
    return out-2.0

gpaDF=data[data.columns[classIndexStart:]].applymap(encode) #apply above encoding function 
gpaDF[data.columns[:classIndexStart]]=data[data.columns[:classIndexStart]]
gpaDF=gpaDF[data.columns]

print gpaDF.head()
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from copy import deepcopy

col = range(classIndexStart,classIndexEnd+1) #all finance classes
pca = PCA(n_components=2)

X_raw = gpaDF[gpaDF.columns[col]]

print "With original data, untransformed"
#X_raw_norm = normalizer.fit_transform(X_raw)
X_raw_norm = deepcopy(X_raw)

reduced_X = pca.fit_transform(X_raw_norm)
y = data['Graduated']

red_x, red_y = [], []
blue_x, blue_y = [], []

for i in range(len(reduced_X)):
    if y[i] == 0:
        red_x.append(reduced_X[i][0])
        red_y.append(reduced_X[i][1])
    elif y[i] == 1:
        blue_x.append(reduced_X[i][0])
        blue_y.append(reduced_X[i][1])

plt.scatter(red_x, red_y, c='r', marker='x',label='Not Graduated')
plt.scatter(blue_x, blue_y, c='b', marker='x',label='Graduated')
plt.legend()

plt.show()
```

```python
def getPCADataFrame(columns, pca, n_components=2):
    
    pcaDF=pd.DataFrame({'class':columns})
    
    for i in range(1,n_components+1):#add principal component columns
        pcaDF['PC'+str(i)]=pca.components_[i-1]
    
    return pcaDF

pcaDF =  getPCADataFrame(gpaDF.columns[col],pca)

print pcaDF

print "PCA Explained variance ratio: ",pca.explained_variance_ratio_

print "\nPCA DF Sorted by PC1"
print pcaDF.sort(columns='PC1')

print "\nPCA DF Sorted by PC2"
print pcaDF.sort(columns='PC2')

print "Total finance major records: ",len(gpaDF)
print "Percentage not graduated: ", 100*round(len(gpaDF.Graduated[gpaDF.Graduated==0])/1.0/len(gpaDF),3)
```

```python
import sys

from sklearn import metrics
import kmeansdistance as KD
from sklearn.cluster import KMeans

col = range(classIndexStart,classIndexEnd+1)

misdistman = []
X = gpaDF[gpaDF.columns[col]]
chIndex = []
silhouettes = []

def CH_index(X, labels, centroids):
    '''
    https://github.com/scampion/scikit-learn/blob/master/scikits/learn/cluster/__init__.py
    '''
    mean = np.mean(X,axis=0) 
    
    B = np.sum([ np.sum(labels==i)*(c - mean)**2 for i,c in enumerate(centroids)])
    W = np.sum([ (x-centroids[labels[i]])**2 
                        for i, x in enumerate(X)])
    c = len(centroids)
    n = len(X)
    return ((n-c)*B )/1.0/((c-1)*W)


for i in range(2,10):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(X)
    kfit = KD.kmeans(X,kmeans.cluster_centers_,metric = "cityblock")
    misdistman.append(sum(np.min(cdist(X, kfit[0],'cityblock'), axis=1)) / X.shape[0])
    if i>1 :
        kmeans_model = KMeans(n_clusters=i).fit(X)
        chIndex.append(CH_index(X.values, kmeans_model.labels_,kmeans.cluster_centers_))
        kfit = KD.kmeans(X,kmeans.cluster_centers_,metric = "cityblock")
        silhouettes.append(metrics.silhouette_score(X,kfit[1],metric='cityblock'))
        print "K=",i," silhouette score", metrics.silhouette_score(X,kfit[1],metric='cityblock')

plt.plot(range(2,10), misdistman, 'bx-', label='meanDist')
plt.xlabel('k')
plt.ylabel('Mean Distortions')
plt.title('KMeans Mean Distortions with L1 metric')
plt.legend()
plt.show()

plt.plot(range(2,10), silhouettes, 'gx-', label= 'silhouette')
plt.xlabel('k')
plt.ylabel('Sil Score')
plt.title('KMeans Silhouette with L1 metric')
plt.legend()
plt.show()

plt.plot(range(2,10), chIndex, 'rx-',label='chIndex')
plt.xlabel('k')
plt.ylabel("L1 metric value")
plt.title("KMeans CH Index")
plt.legend()
plt.show()
```

# Logistic Regression - Important Classes and Coeffs

```python
col = range(classIndexStart,classIndexEnd+1)
X_train, X_test, y_train, y_test = train_test_split(gpaDF[gpaDF.columns[col]], data['Graduated'])

classifier = LogisticRegression()
classifier.fit(X_train.reshape(len(X_train),len(col)), y_train)
y_pred = classifier.predict(X_test.reshape(len(X_test),len(col)))

zippedClassCoef = zip(classifier.coef_[0],gpaDF.columns[col])
zippedClassCoef.sort()
print "\n\n   LogReg Coeff     Class Name"
print '/','-'*28,'\\'
for i in zippedClassCoef[::-1]:
    print '{0:10.5f}{1:10}{2}'.format(i[0],'',i[1])
    print '|','-'*28,'|'
print '\\','-'*28,'/'

print "\n\nNow with absolute values!"

absZippedClassCoef = zip([abs(x) for x in classifier.coef_[0]],gpaDF.columns[col])
absZippedClassCoef.sort()
print "\n\n   Abs LogReg Coeff     Class Name"
print '/','-'*28,'\\'
for i in absZippedClassCoef[::-1]:
    print '{0:10.5f}{1:10}{2}'.format(i[0],'',i[1])
    print '|','-'*28,'|'
print '\\','-'*28,'/'

print "Most Important Classes Normal Val Coefficient: (in order)"
print [x[1] for x in zippedClassCoef[-5:]][::-1]#import classes with regular value list
print "Most Important Classes Abs Val Coefficient: (in order)"
print [x[1] for x in absZippedClassCoef[-5:]][::-1]#import classes with absolute value list

printContingencyTable(y_test,y_pred,['Graduated','LogReg Label'])
scores = cross_val_score(classifier, X_train, y_train, cv=5)
print 'Accuracy:  ', np.mean(scores)
precisions = cross_val_score(classifier, X_train, y_train, cv=5, scoring='precision')
print 'Precision: ', np.mean(precisions)
recalls = cross_val_score(classifier, X_train, y_train, cv=5, scoring='recall')
print 'Recall:    ', np.mean(recalls)
f1s = cross_val_score(classifier, X_train, y_train, cv=5, scoring='f1')
print 'F1:        ', np.mean(f1s)
```

```python
#
#Show that labels are nearly as strong as the full feature set (i.e. analog of an enterotype for student performance)
#

#find clusters
#find percentage of people graduating in each cluster e.g. 0.1, 0.99, 0.4
#replace cluster labels of 0,1 or 0,1,2 with correpsonding cluster graduation rates
#feed that to the ROC

col = range(classIndexStart,classIndexEnd+1)

X_train, X_test, y_train, y_test = train_test_split(gpaDF[gpaDF.columns[col]], gpaDF['Graduated'])


k=2
classifier = KMeans(n_clusters=k, n_init=100)
classifier.fit(X_train)
y_pred = classifier.predict(X_test)

plt.figure(figsize=(9,6))
false_positive_rate, true_positive_rate, thresholds = roc_curve(y_test,y_pred)
roc_auc = auc(false_positive_rate, true_positive_rate)
plt.plot(false_positive_rate, true_positive_rate, 'b--', label='KMeans AUC = %0.2f' %roc_auc)

classifier = LogisticRegression()
classifier.fit(X_train.reshape(len(X_train),len(col)), y_train)
y_pred = classifier.predict_proba(X_test)[:,1]

false_positive_rate, true_positive_rate, thresholds = roc_curve(y_test,y_pred)
roc_auc = auc(false_positive_rate, true_positive_rate)
plt.plot(false_positive_rate, true_positive_rate, 'b', label='LogReg AUC = %0.2f' %roc_auc)
plt.title('ROC Curve for regular 0-1 labeling of KMeans (K=2) and Logistic Regression.')
plt.legend(loc='lower right')
plt.plot([0, 1], [0, 1], 'r--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.ylabel('False Negative Rate')
plt.xlabel('True Positive Rate')
plt.show()



print "Now running KMeans on the FULL dataset, K=2"
k=2
classifier = KMeans(n_clusters=k, n_init=100)
classifier.fit(gpaDF[gpaDF.columns[col]])
y_pred = classifier.predict(gpaDF[gpaDF.columns[col]])

gpaDF['clusterLbl'] = y_pred

gpaDF_CL1 = deepcopy(gpaDF[gpaDF['clusterLbl']==0])
gpaDF_CL1_Perc_Graduating = len(gpaDF_CL1[gpaDF_CL1['Graduated']==1])/1.0/len(gpaDF_CL1)
print "\n\nLength of cluster 1: ",len(gpaDF_CL1)
print "Avg number of terms: ", gpaDF_CL1.N_Terms.mean()
print "Avg number of inst credits: ", gpaDF_CL1.Inst_Cr.mean()
print "Percentage that are graduating: ",gpaDF_CL1_Perc_Graduating

gpaDF_CL2 = deepcopy(gpaDF[gpaDF['clusterLbl']==1])
gpaDF_CL2_Perc_Graduating = len(gpaDF_CL2[gpaDF_CL2['Graduated']==1])/1.0/len(gpaDF_CL2)
print "\n\nLength of cluster 2: ",len(gpaDF_CL2)
print "Avg number of terms: ", gpaDF_CL2.N_Terms.mean()
print "Avg number of inst credits: ", gpaDF_CL2.Inst_Cr.mean()
print "Percentage that are graduating: ",gpaDF_CL2_Perc_Graduating

clusterGradPercMap = {0:gpaDF_CL1_Perc_Graduating,1:gpaDF_CL2_Perc_Graduating}
gpaDF['clusterLblPerc']=gpaDF['clusterLbl'].map(clusterGradPercMap)

plt.figure(figsize=(9,6))
false_positive_rate, true_positive_rate, thresholds = roc_curve(gpaDF['Graduated'],gpaDF['clusterLblPerc'])
roc_auc = auc(false_positive_rate, true_positive_rate)
plt.plot(false_positive_rate, true_positive_rate, 'b', label='ClusterPerc AUC = %0.2f' %roc_auc)
false_positive_rate, true_positive_rate, thresholds = roc_curve(gpaDF['Graduated'],gpaDF['clusterLbl'])
roc_auc = auc(false_positive_rate, true_positive_rate)
plt.plot(false_positive_rate, true_positive_rate, 'g-', label='ClusterBinary AUC = %0.2f' %roc_auc)
plt.title('ROC with k=2 with KMEANS cluster graduating percentage re-labeling.')
plt.legend(loc='lower right')
plt.plot([0, 1], [0, 1], 'r--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.ylabel('False Negative Rate')
plt.xlabel('True Positive Rate')
plt.show()

plt.figure(figsize=(9,9))
plt.subplot(2,1,1)
plt.boxplot([gpaDF_CL1.N_Terms,gpaDF_CL2.N_Terms])
plt.title("N Terms across cluster 1, 2")
plt.subplot(2,1,2)
plt.boxplot([gpaDF_CL1.Inst_Cr,gpaDF_CL2.Inst_Cr])
plt.title("Inst Credit across cluster 1, 2")
plt.show()
```
