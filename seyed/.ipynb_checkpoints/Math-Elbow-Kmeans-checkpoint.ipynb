```python
import numpy as np
%matplotlib inline
import pandas as pd
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from mpl_toolkits.mplot3d import Axes3D
mathdata = pd.read_csv('math-redacted.csv')


student = [[ 0 for _ in range(115)] for _ in range(800)] #array Implementation

for y in range(6,120):
    names = mathdata.columns[y]
    for i in range(0,799):
        student[i][y-6] = mathdata[names][i]
#print x[0][2] this matches the table  

kmeans = KMeans(n_clusters=2)   #for instantiation purposes 
kmeans.fit(student)
#print kmeans.cluster_centers_

K = range(1, 30)
meandistortions = []
''' Because of kmeans valuation this will have a conflict with the next cell 
for k in K:
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(student)
    meandistortions.append(sum(np.min(cdist(student, kmeans.cluster_centers_, 'euclidean'), axis=1)) / len(student) )
plt.plot(K, meandistortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Average distortion')
plt.title('Selecting k with the Elbow Method')
plt.show() '''
```

```python

def findclus(n):    # write a function for n numper of clusters
    cluster = []
    temp = []
    a = 0;
    kmeans = KMeans(n_clusters=n)
    kmeans.fit(student)
    while a < n :
        temp = []
        for std in range(0,799):
            if kmeans.labels_[std] == a :
               temp.append(std)
            else:
               continue 
        a+=1;
        cluster.append(temp)
  
    return cluster


cluster = findclus(2)
print len(cluster[0])   #making sure it works 
#print cluster[i][m]  return index of i ( being the cluster number) and the mth element being the number of element
print cluster[0]
```

```python
# now lets find the percentage of graduates in each cluster given the 'n' number of clusters
grad = 0 ;
# this first part is finding the number of grads in each cluster
def gradperc(n):
    nerds = [];
    for i in range(0,n): 
        grad = 0;
        for j in cluster[i]:
            if mathdata['Graduated'][j] == 'Y':
                 grad+=1
        nerds.append(grad)
#    print nerds
    # this part needs to find out the percentage of grads in their clusters
    percentage = [];
    for i in range(0,n):
        percentage.append((float(nerds[i]) / len(cluster[i]))*100)
    # find the biggest percentage
    for i in range (0,n):
        for j in range (0,n-1):
            if percentage[j+1]> percentage[j]:
                percentage[j] = percentage[j+1]
    
    return percentage[0]

#print gradperc(2)   for testing purposes

yax = []
# finally , we have to compare the percentages for the greatest value amongs different number of clusters
# graphin would make it nice
for i in range(2,100):
    cluster = findclus(i)
    yax.append(gradperc(i))

plt.plot(range(2,100),yax, 'bx-')
plt.xlabel('number of clusters')
plt.ylabel('max percentage of graduates found in K clusters ')
plt.title('isolating our graduates')
plt.show()
```

```python
kmeans = KMeans(n_clusters=2)   #for instantiation purposes 
kmeans.fit(student)
cluster0 = []
cluster1 = []
for count in range(0,799):
    if kmeans.labels_[count] == 0:
        cluster0.append(count)
    else:
        cluster1.append(count)

#print len(cluster0) just making sure they add up to 800
#print len(cluster1)

grad0 = 0 ;
grad1 = 0;
for i in cluster0:
        if mathdata['Graduated'][i] == 'Y':
            grad0+=1
for i in cluster1:
        if mathdata['Graduated'][i] == 'Y':
            grad1+=1

            
# Illustration purposes

print 'number of grads in 1st cluster : ' + str(grad0)  # yes !  this is all the nerds
print 'number of grads in 2nd cluster : ' + str(grad1)   # Yes ! yes !! !!! ! that is 97% accuracy! 

print 'number of students in 1st cluster : ' + str(len(cluster0))  # yes ! 
print 'number of students in 2nd cluster : ' + str(len(cluster1))   # Yes ! yes !! !!! ! that is 97% accuracy! 

print str((float(grad0)/len(cluster0))*100) + ' percentage of grads in 1st cluster' 
print str((float(grad1)/len(cluster1))*100) + ' percentage of grads in 2nd cluster'
```

```python
#now lets try and go beyond this 2 clusters to see any pattern is found 
# meanwhile , I am NOT removing the 1st cluster of grads out of the dataset
kmeans = KMeans(n_clusters=3)   #for instantiation purposes 
kmeans.fit(student)


cluster0 = []
cluster1 = []
cluster2 = []
for count in range(0,799):
    if kmeans.labels_[count] == 0:
        cluster0.append(count)
    elif kmeans.labels_[count] == 1:
        cluster1.append(count)
    elif kmeans.labels_[count] == 2:
        cluster2.append(count)

      
grad0 = 0 ;
grad1 = 0;
grad2 = 0 ; 
for i in cluster0:
        if mathdata['Graduated'][i] == 'Y':
            grad0+=1
for i in cluster1:
        if mathdata['Graduated'][i] == 'Y':
            grad1+=1
for i in cluster2:
        if mathdata['Graduated'][i] == 'Y':
            grad2+=1
       
# Illustration purposes

print 'number of grads in 1st cluster : ' + str(grad0)  # no one graduated here ! :(
print 'number of grads in 2nd cluster : ' + str(grad1)   
print 'number of grads in 3rd cluster : ' + str(grad2)  # Interestingly same # of grads are in this guy

print 'number of students in 1st cluster : ' + str(len(cluster0))  # yes ! 
print 'number of students in 2nd cluster : ' + str(len(cluster1))   # making sure they add up correctly 
print 'number of students in 3rd cluster : ' + str(len(cluster2))   

print str((float(grad0)/len(cluster0))*100) + ' percentage of grads in 1st cluster' 
print str((float(grad1)/len(cluster1))*100) + ' percentage of grads in 2nd cluster'
print str((float(grad2)/len(cluster2))*100) + ' percentage of grads in 3rd cluster'
```
